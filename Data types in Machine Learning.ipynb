```python
import tensorflow as tf
string = tf.Variable([['this is a string' ,'v'],['this is me','t']], tf.string)
number = tf.Variable(323, tf.int16)
print(string)
print(number)
tf.rank(string)
string.shape
```

```python
tensor_1 = tf.ones([1, 2, 3])
print(tensor_1)
tensor_2 = tf.reshape(tensor_1,[2, 3, 1])
print(tensor_2)
tensor_3 = tf.reshape(tensor_2, [3, -1])
print(tensor_3)
```

```python
t = tf.zeros([5,5,5,5])
print(t)
t = tf.reshape(t, [25, -1])
print(t)
```
